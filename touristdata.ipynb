```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data into Pandas DataFrame
df = pd.read_csv('Tourist survey_compile data.csv', encoding='latin-1')

# Gender pie chart
plt.figure() 
gender = df['Gender'].value_counts()
labels = gender.index
sizes = gender.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Gender')
plt.savefig('gender.jpeg',dpi = 300)
# Visiting places bar chart  
plt.figure()
top_places = df['Visiting Place'].value_counts()[:5]
plt.bar(top_places.index, top_places.values)
plt.xticks(rotation=45)
plt.title('Top 5 Visiting Places') 

# Age histogram
plt.figure() 
plt.hist(df['Age'], bins=10)
plt.title('Age Distribution')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Set plotting style 
plt.style.use('ggplot') 

# Custom colors
colors = ['#581845', '#900C3F', '#C70039']

# Gender pie chart
plt.figure()
df = pd.read_csv('Tourist survey_compile data.csv', encoding='latin-1')
gender = df['Gender'].value_counts()
plt.pie(gender, colors=colors, shadow=True, startangle=140)
plt.title('Gender', fontsize=16)

# Visiting places bar chart
plt.figure(figsize=(8,6)) 
top_places = df['Visiting Place'].value_counts()[:5]
plt.bar(top_places.index, top_places, color=colors[0])
plt.xticks(rotation=45, fontsize=12)
plt.ylabel('Number of Visitors', fontsize=12) 
plt.title('Top 5 Visiting Places', fontsize=16)
plt.savefig('visit.jpeg',dpi = 300,bbox_inches ='tight')

# Age histogram 
plt.figure(figsize=(8,4))
plt.hist(df['Age'], edgecolor='black', bins=5)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12) 
plt.title('Age Distribution', fontsize=16)
plt.savefig('age.jpeg',dpi = 300, bbox ='tight')
# # Tight layout and show
# plt.tight_layout()
# plt.show()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
sns.displot(data=df, x="Age",bins = 7, kde = True)
plt.ylabel('Frequency')
plt.savefig('age_dis.jpeg',dpi = 300, bbox_inches ='tight')
```

```python
sns.set(font_scale=1.4) 
sns.set_style("whitegrid")
sns.catplot(data=df, x="Age", y="Gender",kind="violin", color=".9", inner=None)
sns.swarmplot(data=df, x="Age", y="Gender", size=7)
plt.savefig('age_class.jpeg',dpi = 300, bbox_inches ='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Tourist survey_compile data.csv', encoding='latin-1')
plt.style.use('ggplot') 

concern_counts = df['Concern about degradation'].value_counts()

plt.figure()
sns.barplot(concern_counts.index, concern_counts.values)
plt.xticks(rotation=45)
plt.xlabel('Level of Concern')
plt.ylabel('Number of Respondents')
plt.title('Concern About Environmental Degradation')
plt.savefig('concern2.jpeg',dpi = 300,bbox_inches = 'tight')
plt.show()
```
